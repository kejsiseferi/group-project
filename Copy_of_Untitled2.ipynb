```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

# 1. FONTI DHE MADHËSIA
mpl.rcParams['font.family'] = 'DejaVu Serif'
mpl.rcParams['font.size'] = 13

# 2. TË DHËNAT - dy funksionet matematikore
t = np.linspace(0, 10, 1000)
x1 = np.sin(t)
x2 = np.sin(t) + np.cos(2*t)

# 3. FIGURA
fig, ax = plt.subplots(figsize=(10, 5))

#ax.set_facecolor('#F8F9FA')
import matplotlib.image as mpimg
img = mpimg.imread('jumbo.jpg')
ax.imshow(img, extent=[0, 10, -2.5, 2.5], aspect='auto', alpha=0.45, zorder=0)

# 4. VIZATIMI I VIJAVE me ngjyra, trashësi dhe markera
ax.plot(t, x1, color='#FF0000', linewidth=6, alpha=0.2, zorder=2)
ax.plot(t, x1,
        color='#FF3366',
        linewidth=2.5,
        marker='o',
        markevery=80,
        markersize=6,
        label=r'$f(t) = \sin(t)$', zorder=3)
#Vija blu
ax.plot(t, x2, color='#006994', linewidth=6, alpha=0.2, zorder=2)
ax.plot(t, x2,
        color='#006994',
        linewidth=2.5,
        marker='s',
        markevery=80,
        markersize=6,
        label=r'$f(t) = \sin(t) + \cos(2t)$', zorder=3)

# 5. STILI I RRJETËS (grid)
ax.grid(True,
        linestyle=':',         # me pika
        linewidth=0.8,
        color='gray',
        alpha=0.6)             # transparencë 60%

# 6. NGJYRAT E SFONDIT
ax.set_facecolor('#F8F9FA')    # sfond i grafikut
fig.patch.set_facecolor('#FFFFFF')  # sfond i figurës

# 7. ETIKETAT DHE TITULLI
ax.set_xlabel('t', fontsize=15, fontweight='bold', color='#1D3557')
ax.set_ylabel('x(t)', fontsize=15, fontweight='bold', color='#1D3557')
ax.set_title('Group Signature Plot', fontsize=18, fontweight='bold',
             color='#1D3557', pad=15)

# 8. LEGJENDA
ax.legend(fontsize=12, loc='upper right', framealpha=0.9, edgecolor='#1D3557')

# 9. RUAJTJA
plt.tight_layout()
plt.show()
plt.savefig('signature_plot.png', dpi=150, bbox_inches='tight')
```
